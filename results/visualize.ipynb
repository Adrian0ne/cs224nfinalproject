```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
albert_pt = pd.read_csv('albert-base-v2.csv')
albert_ft = pd.read_csv('albert-base-v2-squad2.csv')
bert_pt = pd.read_csv('bert-base-uncased.csv')
bert_ft = pd.read_csv('bert-base-uncased-squad2.csv')
```

```python
# Add model column to each dataframe
albert_pt['model'] = 'ALBERT Pretrained'
albert_ft['model'] = 'ALBERT Fine-tuned'
bert_pt['model'] = 'BERT Pretrained'
bert_ft['model'] = 'BERT Fine-tuned'
```

```python
# Create a new dataframe, and append e
df = pd.DataFrame(None, columns=albert_pt.columns)
df = df.append(albert_pt, ignore_index=True)
df = df.append(albert_ft, ignore_index=True)
df = df.append(bert_pt, ignore_index=True)
df = df.append(bert_ft, ignore_index=True)
```

```python
df.head()
```

```python
heatmap_df = df.pivot("model", "layer", "f1")
sns.heatmap(data=heatmap_df, annot=True, linewidths=.5, center=48)
plt.xlabel("")
plt.ylabel("");
```

```python
heatmap_df = df.pivot("model", "layer", "exact")
sns.heatmap(data=heatmap_df, annot=True, linewidths=.5, center=48)
plt.xlabel("")
plt.ylabel("");
```
